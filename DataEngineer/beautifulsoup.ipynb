# web scrap crypto news

```python
url = 'https://cryptonews.com/'
source = requests.get(url).text
soup = BeautifulSoup(source, 'lxml')
COL = ['category', 'title', 'url', 'post time']
result = list()
```

```python
article = soup.find('div', class_='cn-tile article')
#print(article.prettify())
# title: article.h4.text
# urrl: 'https://' + article.a['href']
```

```python
for article in soup.find_all('div', attrs={'class': ['cn-tile article', 'cn-tile row article']}):
    try:
        title = article.h4.text
    except Exception as e:
        title = None
    try:
        site = 'https://' + article.a['href']
    except Exception as e:
        site = None
    postTime = article.time.text
    category = article.div.a.text
    result = result + [[category, title, site, postTime]]
result=pd.DataFrame(result, columns=COL).sort_values('category')\
        .reset_index(drop=True)
result
```
